```python
import pandas as pd

dataset_columns=["polarity","ids","date","flag","user","text"]
dataset_encoding="ISO-8859-1"

#This is the sentiment140 dataset containing 1.6M tweets
#The tweets have been annotated (0 = negative, 2 = neutral, 4 = positive)
#Fields:
# target: the polarity of the tweet (0 = negative, 2 = neutral, 4 = positive)
# ids: The id of the tweet ( 2087)
# date: the date of the tweet (Sat May 16 23:58:44 UTC 2009)
# flag: The query (lyx). If there is no query, then this value is NO_QUERY.
# user: the user that tweeted (robotickilldozr)
# text: the text of the tweet (Lyx is cool)
dataset=pd.read_csv("sentiment140Tweets.csv", 
                    names=dataset_columns, 
                    encoding=dataset_encoding)
```

```python
#Select the polarity and text columns
x=dataset.iloc[:,[0,5]]
x.head()
```

```python
#Cleaning the data

#Decoding HTML
from bs4 import BeautifulSoup
x["text"]=x["text"].map(lambda x: BeautifulSoup(x,"lxml").get_text())

#Removing "@" mentions
import re
x["text"]=x["text"].map(lambda x: re.sub("@\S+"," ", x))

#Removing links
x["text"]=x["text"].map(lambda x: re.sub("https?:\S+|http?:\S|[^A-Za-z0-9]+|www.[^ ]+"," ", x))

#Removing numbers
x["text"]=x["text"].map(lambda x: re.sub("\d+"," ", x))

#Convert to lowercase
import string
x["text"]=x["text"].map(lambda x: x.lower())

#Tokenize posts text into words
from nltk.tokenize import word_tokenize
x["text"]=x["text"].map(lambda x: word_tokenize(x))

#Remove punctuation
x["text"]=x["text"].map(lambda x:[word for word in x if word not in string.punctuation])

#Remove stop words
from nltk.corpus import stopwords
x["text"]=x["text"].map(lambda x:[word for word in x if word not in stopwords.words("english")])

#Lemmatize words to convert them into base form
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
x["text"]=x["text"].map(lambda x: [lemmatizer.lemmatize(word) for word in x])

#Convert tokens back into a string
x["text"]=x["text"].map(lambda x: " ".join(x))

x.head()
```

```python
print("Done cleaning")
```

```python
#Export cleaned data to a new CSV file
x.to_csv("sentiment140CleanNOSW.csv")
```
